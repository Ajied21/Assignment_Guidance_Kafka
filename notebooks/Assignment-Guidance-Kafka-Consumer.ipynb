## Kafka Consumer

```python
import json
import uuid
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from kafka import KafkaConsumer
from kafka import TopicPartition
from faker import Faker
from time import sleep
```

```python
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)
```

```python
kafka_host = os.getenv('KAFKA_HOST')
```

```python
kafka_topic_partition = "Data_Transaksi_Keuangan_E-commerce"
```

```python
# To consume latest messages and auto-commit offsets

consumer = KafkaConsumer(kafka_topic_partition,
                         group_id='consumer-group_0',
                         bootstrap_servers=[f'{kafka_host}:9092'])
```

```python

count = 0  # Inisialisasi penghitung pesan

for message in consumer:
    print("nomor pesan ke:", count)
    # Cetak detail pesan
    print("%s:%d:%d: key=%s value=%s" % (message.topic, message.partition,
                                         message.offset, message.key,
                                         message.value))
    
    # Tambah penghitung setiap kali pesan diterima
    count += 1
    # Cetak jumlah total pesan setelah selesai (atau di bagian lain sesuai kebutuhan)
    print("\n")
```

```python
consumer = KafkaConsumer(group_id='consumer-group',
                         bootstrap_servers=[f'{kafka_host}:9092'],
                         value_deserializer=lambda m: json.loads(m.decode('ascii')),
                         auto_offset_reset='earliest',
                         enable_auto_commit=False
                        )
```

## Save data to json

```python
#Transaksi_Keuangan_E_commerce
# Konfigurasi Kafka
consumer.assign([
    TopicPartition(kafka_topic_partition, 0),
    TopicPartition(kafka_topic_partition, 1),
    TopicPartition(kafka_topic_partition, 2),
    TopicPartition(kafka_topic_partition, 3),
    TopicPartition(kafka_topic_partition, 4)
])

consumer.seek_to_beginning()

# Direktori dan file untuk menyimpan data
#json_directory = "/resource/KSQLDB_Kafka/Data"
#json_file_path = os.path.join(json_directory, "Data_Transaksi_Keuangan_E_commerce.json")

# Proses pesan dari Kafka
for message in consumer:
    print("%s:%d:%d: key=%s value=%s" % (message.topic, message.partition,
                                         message.offset, message.key,
                                         message.value))
    print("\n")

# Menyimpan pesan sebagai JSON
    try:
        # Cek apakah message.value berupa bytes atau dict
        if isinstance(message.value, bytes):
            value = message.value.decode('utf-8')  # Decode jika tipe data bytes
        else:
            value = message.value  # Jika sudah berupa dict atau tipe lain, langsung gunakan

        # Simpan pesan sebagai JSON
        with open("Transaksi_Keuangan_E_commerce.json", 'w') as json_file:
            json.dump({
                "topic": message.topic,
                "partition": message.partition,
                "offset": message.offset,
                "key": message.key.decode('utf-8') if message.key else None,
                "value": value
            }, json_file, indent=4)
        
        print("Data berhasil disimpan ke JSON.")
    except Exception as e:
        print(f"Gagal menyimpan data ke JSON karena: {e}")
```
